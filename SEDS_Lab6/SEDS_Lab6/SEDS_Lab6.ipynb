````python
# remove unnecessary punctuation, tags
import string

text = """
Data science incorporates tools from multiple disciplines to gather a data set, process, and derive insights from the data set, extract meaningful data from the set, and interpret it for decision-making purposes. 
The disciplinary areas that make up the data science field include mining, statistics, machine learning, analytics, and programming.

Data mining applies algorithms to the complex data set to reveal patterns that are then used to extract useful and relevant data from the set. 
Statistical measures or predictive analytics use this extracted data to gauge events that are likely to happen in the future based on what the data shows happened in the past.

Machine learning is an artificial intelligence tool that processes mass quantities of data that a human would be unable to process in a lifetime. Machine learning perfects the decision model presented under predictive analytics by matching the likelihood of an event happening to what actually happened at a predicted time.
Using analytics, the data analyst collects and processes the structured data from the machine learning stage using algorithms. 
The analyst interprets, converts, and summarizes the data into a cohesive language that the decision-making team can understand.
Data science is applied to practically all contexts and, as the data scientist's role evolves, the field will expand to encompass data architecture, data engineering, and data administration.
"""

translator = str.maketrans('', '', string.punctuation + string.digits)
text = text.translate(translator).lower()

print(text)    
````

````python
# tokenization:Breaking down the text into individual words or tokens
import nltk
nltk.download('punkt_tab')
nltk.download('punkt')

tokens = nltk.word_tokenize(text)
print(tokens)
````

````python
# Remove stop words:
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
en_stopwords = stopwords.words('english')
en_stopwords = set(en_stopwords)
tokens = [w for w in tokens if w not in en_stopwords]
print(tokens)
````

````python
# Word frequency count:
cnt_fdist = nltk.FreqDist(tokens)
cnt_fdist.most_common(20)
````

````python
import matplotlib.pyplot as plt

common_words = [word[0] for word in cnt_fdist.most_common(20)]
word_freq = [freq[1] for freq in cnt_fdist.most_common(20)]

plt.bar(common_words, word_freq)
plt.title('Top 20 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=80)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
````

````python
from wordcloud import WordCloud
wordcloud = WordCloud(collocations=False).generate(' '.join(tokens))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
````

````python
# challenge 3:
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import lxml
import string
import nltk
from nltk.corpus import stopwords

url = 'https://en.wikipedia.org/wiki/Data_science'
page = urlopen(url).read().decode('utf-8')
soup = bs(page) # this do like all the html and xml tags are here.

links = soup.find_all('a') # it will generate like an array with all 'a' elements
all_link_text = []

all_link_text.extend([a.text for a in links])

text = ' '.join(all_link_text)
# remove unesasary punctations
translator = str.maketrans('\n', ' ', string.punctuation + string.digits)
text = text.translate(translator).lower()
#tokenization
tokens = nltk.word_tokenize(text)
# Removing stop words
nltk.download('stopwords')
en_stopwords = stopwords.words('english')
en_stopwords = set(en_stopwords)
words = [w for w in tokens if w not in en_stopwords]

# frequently used
cnt_fdist = nltk.FreqDist(words)
common_words = [word[0] for word in cnt_fdist.most_common(20)]
word_freq = [freq[1] for freq in cnt_fdist.most_common(20)]
# plotting 
plt.bar(common_words, word_freq)
plt.title('Top 20 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=80)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

#creating card
wordcloud = WordCloud(collocations=False).generate(' '.join(words)) # generate get a long string with words separated by space and calc the frequency of each word
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
````

````python
from urllib.request import urlopen

text = """
Data science incorporates tools from multiple disciplines to gather a data set, process, and derive insights from the data set, extract meaningful data from the set, and interpret it for decision-making purposes. 
The disciplinary areas that make up the data science field include mining, statistics, machine learning, analytics, and programming.

Data mining applies algorithms to the complex data set to reveal patterns that are then used to extract useful and relevant data from the set. 
Statistical measures or predictive analytics use this extracted data to gauge events that are likely to happen in the future based on what the data shows happened in the past.

Machine learning is an artificial intelligence tool that processes mass quantities of data that a human would be unable to process in a lifetime. Machine learning perfects the decision model presented under predictive analytics by matching the likelihood of an event happening to what actually happened at a predicted time.
Using analytics, the data analyst collects and processes the structured data from the machine learning stage using algorithms. 
The analyst interprets, converts, and summarizes the data into a cohesive language that the decision-making team can understand.
Data science is applied to practically all contexts and, as the data scientist's role evolves, the field will expand to encompass data architecture, data engineering, and data administration.
"""

# do all links in one list
url_list = list()
url_list.append('https://www.heavy.ai/learn/data-science')
url_list.append('https://en.wikipedia.org/wiki/Data_science')
url_list.append('https://www.ibm.com/cloud/learn/data-science-introduction')
url_list.append('https://deepai.org/machine-learning-glossary-and-terms/data-science')

# in each link get just p elements and join them with the first text
for url in url_list:
    page = urlopen(url).read().decode('utf-8')
    soup = bs(page)
    links = soup.find_all('p')
    all_link_text = []
    all_link_text.extend([p.text for p in links])
    text += ''.join(all_link_text)

#remove unessesary data 
translator = str.maketrans('\n', ' ', string.punctuation + string.digits+ '“' + '”' + '-' + '–' + '’')
text = text.translate(translator).lower()

#tokenization
nltk.download('punkt')
tokens = nltk.word_tokenize(text)

#remove stop words...etc
nltk.download('stopwords')
en_stopwords = stopwords.words('english')
en_stopwords = set(en_stopwords)
words = [w for w in tokens if w not in en_stopwords]

#calc the most common values
cnt_fdist = nltk.FreqDist(words)
common_words = [word[0] for word in cnt_fdist.most_common(20)]
word_freq = [freq[1] for freq in cnt_fdist.most_common(20)]

# plotting 
plt.bar(common_words, word_freq)
plt.title('Top 20 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=80)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

#creating card
wordcloud = WordCloud(collocations=False).generate(' '.join(words)) # generate get a long string with words separated by space and calc the frequency of each word
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
````

````python
from bs4 import BeautifulSoup as bs
import lxml
from urllib.request import urlopen
import string
import pandas as pd
````

````python
# take the links and extract the text from
url_list = list()
url_list.append('https://www.goodreads.com/quotes/tag/motivational?page=1')
url_list.append('https://www.goodreads.com/quotes/tag/motivational?page=2')
url_list.append('https://www.goodreads.com/quotes/tag/motivational?page=3')
url_list.append('https://www.goodreads.com/quotes/tag/motivational?page=4')
url_list.append('https://www.goodreads.com/quotes/tag/motivational?page=5')
````

````python
text = ""
data = []  # To store text from the page
tags_list = []
# Get data from pages
for link in url_list:
    page = urlopen(link).read().decode('utf-8')
    soup = bs(page, 'html.parser')  # Specify parser
    
    # Find all divs
    quotes = soup.find_all("div", {"class": "quoteDetails"}) # quotes is a list of each div.quoteDetails element that have all html nodes inside

    # clean text , and get author from each quote
    for quote in quotes:
        quote_text_div = quote.find("div", {"class": "quoteText"}) # give a list of text
        if quote_text_div:
            author = quote_text_div.find("span",{"class":"authorOrTitle"}).get_text(strip=True)[:-1]
            book = quote_text_div.find("a",{"class":"authorOrTitle"}).get_text(strip=True)if quote.find('a',{"class":"authorOrTitle"}) else "Unkown"

            text = quote_text_div.get_text(strip=True)
            text = text.replace(author, "").replace(book, "").strip()
            translator = str.maketrans('\n', ' ', string.punctuation + string.digits+ '“' + '”' + '–' + '―')
            text = text.translate(translator).lower()
        
        
        quote_footer_div = quote.find("div", {"class": "quoteFooter"})
        tags = quote_footer_div.find("div", {"class": "greyText"})
        tags = tags.get_text(strip=True)[5:]
        tags_list.extend(tags.split(','))
        likes = quote_footer_div.find("a", {"title": "View this quote"}).get_text(strip=True)[:-6]

        
    
        data.append({
                    'Text': text,
                    'Author': author,
                    'Book':book,
                    'Tags' : tags,
                    'Likes' : likes
                })       
    
df = pd.DataFrame(data)
df
````

````python
#creating card
wordcloud = WordCloud(collocations=False).generate(' '.join(words)) # generate get a long string with words separated by space and calc the frequency of each word
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
````
